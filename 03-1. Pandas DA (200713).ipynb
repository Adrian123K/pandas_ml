## <b>■ 유방암 데이터 knn 을 파이썬으로 구현하기</b>

```python
import pandas as pd
import seaborn as sns
from sklearn import preprocessing # sklearn 사용
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split # 훈련데이터와 테스트 데이터를 분리하기 위한 패키지
from sklearn import metrics # 이원교차표를 보기위한 패키지
from sklearn.metrics import accuracy_score

df = pd.read_csv("wisc_bc_data.csv")  # 1장 내용
dia = pd.get_dummies(df['diagnosis'],prefix='diagnosis') # 4장 더미변수 생성 : 두 개의 더미변수가 생성
# diagnosis_B : 종양이 양성(B)이면 1, 아니면 0으로 출력
# diagnosis_M : 종양이 악성(M)이면 1, 아니면 0으로 출력

df = pd.concat([df, dia],axis=1)
df.drop (['diagnosis'],axis=1,inplace=True)

X = df.iloc[:,:-2].values #diagnosis_B도 제거. 훈련데이터 준비
y = df['diagnosis_M'] # 라벨 준비

X=preprocessing.StandardScaler().fit(X).transform(X) # 훈련데이터 정규화

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size =0.3, random_state = 10) # 훈련데이터 : 테스트 데이터 = 7:3

knn = KNeighborsClassifier(n_neighbors =26) #k=26 노가다로 알아내야한다.
knn.fit(X_train, y_train) # 훈련 코드

y_hat = knn.predict(X_test) # 테스트 데이터의 성능 확인을 위한 예측 코드

knn_matrix = metrics.confusion_matrix(y_test,y_hat) # 이원 교차표
print(knn_matrix)

accuracy = accuracy_score( y_test, y_hat) # 정확도 확인
print(accuracy)
```

### <b>■ read_csv로 데이터프레임 구현하기</b>
    문법
        df = pd.read_csv(file_path, 옵션)
    
    옵션
        path : 파일의 위치, url
        sep : 텍스트 데이터를 필드별로 구분하는 문자
        header : column명 표시 여부
        index_col : row index로 사용할 문자열 리스트
        names : column명으로 사용할 문자열 리스트
        encoding : 텍스트 인코딩의 종류를 지정
        ※ 파이썬에서 한글이 깨지지 않으려면 encoding='utf-8'로 지정
