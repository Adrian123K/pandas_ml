#### <b>※ 확률밀도함수 그래프 설명</b>
    확률밀도함수란 확률 변수의 분포를 나타내는 함수이다.
    확률변수는 크게 2가지
        1. 이산확률변수 : 정수와 같이 명확한 값을 변수 값으로 함
        2. 연속확률변수 : 변수 값이 정수처럼 명확하지 못함. 확률변수가 연속량으로 표기되어 가능한 변수 값의 개수를 셀 수 없는 변수
        
        
$$f(x)={{1}\over {\sqrt{2 \pi \sigma^2}}}\exp ^{-{(x-m)^{2}}\over {2 \sigma ^2}}$$

### ※ 문제137. 확률밀도함수 공식을 이용하지 않고 초등학생 10만 명의 키를 랜덤으로 생성해서 다음의 문제를 푸시오.
### 10만 명의 초등학생 중  무작위로 한 명을 추출했을 때 이 어린이의 키가 145~150 사이에 있을 확률은?
    np.random.randn(100_000)*5 + 140

```python
import numpy as np
height = np.random.randn(100_000)*5 + 140
cnt = 0
for i in height:
    if i>=145 and i<=150:
        cnt += 1
cnt/100000
```

### ※ 문제138. 위의 초등학생 키 데이터로 확률밀도함수 그래프를 그리시오. 
    초등학생 키 140~160 사이의 키로 무작위로 생성하고 평균키를 150, 표준편차를 5로해서 확률밀도함수 그래프를 그리시오
    패키지 scipy 이용

```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt

height = np.arange(140,160,0.001)
plt.plot(height, norm.pdf(height, 150, 5))
plt.show()
```
