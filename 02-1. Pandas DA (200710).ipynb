### ※ 문제40. 접촉자수(contact_number)가 70명이 넘는 환자들의 모든 컬럼을 출력하시오(PatientInfo.csv)

````python
import pandas as pd
pf = pd.read_csv("PatientInfo.csv")
a = list(pf.contact_number.fillna('0'))
for i in range(len(a)) :
    if a[i] == '-' :
        a[i] = 0
list_a = list(map(int,a))
pf['contact'] = list_a
pf[pf.contact>=70]
````

### ※ 문제53. Case.csv에서 도시(City)를 출력하고 도시별 확진자수의 토탈값을 출력하시오

````python
import pandas as pd
case = pd.read_csv('Case.csv')
rs = case.groupby('province')['confirmed'].sum()
rs_df = pd.DataFrame({'city':rs.index,'total':rs.values})
rs_df.sort_values(by='total',ascending=False)
````
