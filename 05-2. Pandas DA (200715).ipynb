### <b>■ 산포도 그래프</b>
    산포도 그래프는 데이터의 퍼짐 정도를 한 눈에 확인할 수 있는 그래프
    이전까지 그린 산포도 그래프는 하나의 범주에 대해서만 작도
        x축 평수, y축 집 값 이면 평수 하나에 대해서만 작도
    seaborn의 산포도를 이용하면 여러개의 범주에 대해서 시각화가 가능

#### <b> 타이타닉호의 클래스 별로 나이의 분포가 어떻게 되는지 산포도 그래프로 확인하시오 (범주형 산포도)</b>

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['figure.figsize']=(30,10)
plt.rcParams.update({'font.size':22})

tit = sns.load_dataset('titanic')
sns.set_style('whitegrid')

sns.stripplot(x='class',y='age',data=tit)
```

#### <b> first class의 승객이 다른 class 보다 나이가 많은 사람이 보이고 first class의 승객 중 나이가 어린 승객이 다른 class의 승객들보다 숫자가 적다</b>
    stripplot 그래프는 분산을 고려하지 않고 시각화
    산포도 그래프의 장점 중 하나가 분산정도(데이터의 퍼짐 정도)를 확인할 수 있는 것이므로 분산까지 같이 보려면 swarmplot 사용

```python
sns.swarmplot(x='class', y='age',data=tit)
```

### <b>■ 막대 그래프</b>
    seaborn의 막대 그래프가 다른 막대 그래프와의 차이점은 hue 옵션이 있어서 막대 그래프를 범주형으로 시각화 할 수 있다.
        예) 3개의 컬럼을 사용
        1. x축 : 성별
        2. y축 : 생존여부의 비율
        3. hue : class(1,2,3등석)

```python
import seaborn as sns

tit = sns.load_dataset('titanic')
sns.set_style('whitegrid')
sns.barplot(x='sex',y='survived',data=tit)
```

```python
sns.barplot(x='sex',y='survived',hue='class',data=tit,) # hue 옵션
```

```python
sns.barplot(x='sex',y='survived',hue='class',dodge=False,data=tit)
```

### ※ 문제146. 사원 테이블의 데이터로 위의 막대 그래프를 그리는데 $x$축을 부서번호, $y$축을 월급으로 하고 hue를 직업으로 해서 시각화 하시오
    부서번호마다 평균으로 어느 직업 월급이 높은지 확인하시오

```python
import seaborn as sns
import pandas as pd

emp = pd.read_csv('emp.csv')
avg_sal = emp.groupby(['deptno','job'])['sal'].mean()
print(avg_sal)
# emp['avg_sal']=

sns.barplot(x='deptno',y='sal',hue='job',data=emp)
```

### <b>■ 바이올린 그래프 (분산확인)</b>
    박스 그래프와 바이올린 그래프는 데이터의 분포와 분산정도를 한 눈에 확인할 때 유용
    박스 그래프만 가지고 데이터의 분산 정도를 정확하게 파악하기 어려울 때 바이올린 그래프를 사용하면 유용

```python
import seaborn as sns
import pandas as pd

tit=sns.load_dataset('titanic')
sns.boxplot(x='alive',y='age',data=tit)
```

```python
sns.boxplot(x='alive',y='age',hue='sex',data=tit)
```

```python
sns.violinplot(x='alive',y='age',data=tit)
```

```python
sns.violinplot(x='alive',y='age',hue='sex',data=tit)
```
