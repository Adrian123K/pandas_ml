### <b>■ 누락 데이터 처리</b>
    머신러닝 데이터 분석의 정확도는 분석 데이터의 품질에 의해 좌우. 데이터 품질을 높이기 위해서는 누락 데이터 처리, 중복 데이터 처리 등 
    오류를 수정하고 분석 목적에 맞게 변형하는 과정이 필요
    
    데이터프레임에는 원소 데이터 값이 종종 누락되는 경우가 존재
    일반적으로 데이터 값이 존재하지 않는 누락 데이터를 NaN으로 표시
    머신러닝 분석 모형에 데이터를 입력하기 전에 반드시 누락 데이터를 제거하거나 다른 적절한 값으로 대체하는 과정이 필요
    누락 데이터가 많아지면 데이터의 품질이 떨어지고 머신러닝 분석 알고리즘을 왜곡하는 현상이 발생
    
    * 누락 데이터 찾는 Pandas method
        1. isnull() : 누락 데이터면 True를 반환
        2. notnull() : 누락 데이터가 아니면 True를 반환

            mean_age=df['age'].mean()

            df['age'].fillna(most_freq, inplace=True)

### ※ 문제153. 타이타닉호의 생존자를 예측하는 머신러닝 모델을 생성하시오(knn 모델)

```python
# ■ 타이타닉 데이터를 knn 으로 수행했을때 
# 1. 필요한 패키지를 import 한다. 
import  pandas  as  pd
import  seaborn  as  sns 

#1단계.  csv ---> 데이터 프레임으로 변환 
df = sns.load_dataset('titanic')

# 컬럼이 모두다 출력될 수 있도록 출력할 열의 개수 한도를 늘리기 
pd.set_option('display.max_columns', 15 ) 
print  (df.head()  )
print ( df.info()  ) 
print ( '\n' )  
print ( df.isnull().sum( axis=0)  )  
```

```python
rdf =  df.drop(['deck', 'embark_town'] , axis= 1) # 누락 데이터가 많은 column을 drop
print ( rdf.columns.values ) 
```

```python
rdf = rdf.dropna( subset=['age'], how='any', axis= 0 ) # 나이의 누락 데이터의 행을 삭제
print ( len(rdf)  ) 
print ('\n')
```

```python
most_freq = rdf['embarked'].value_counts(dropna=True).idxmax() # 정박한 항구의 누락 데이터 2건을 최빈값으로 치환
rdf['embarked'].fillna( most_freq, inplace=True ) 
print ( rdf.isnull().sum(axis=0)  ) 
```

```python
ndf = rdf[ ['survived', 'pclass', 'sex', 'age', 'sibsp', 'parch', 'embarked']] 
print ( ndf.head()  )

gender = pd.get_dummies( ndf['sex'] ) # 명목형 데이터를 숫자로 변환하기 위해 더미변수를 생성
ndf = pd.concat( [ndf, gender], axis=1 ) 
print(ndf.head())
```

```python
onehot_embarked = pd.get_dummies( ndf['embarked'], prefix='town') # 명목형 데이터를 숫자로 변환하기 위해서 더미변수 생성
ndf = pd.concat([ ndf, onehot_embarked], axis=1)
print(ndf.head())
```

```python
ndf.drop( ['sex', 'embarked'],  axis=1, inplace=True ) # 명목형 column 2개를 삭제
print ( ndf.head() ) 
```

```python
# 훈련할 학습 데이터 X와 정답이 있는 라벨 y를 생성
X = ndf[ [ 'pclass' ,'age', 'sibsp','parch', 'female', 'male', 'town_C', 'town_Q',  'town_S']]
y = ndf['survived']   # 종속변수 
y.value_counts() # 0이 생존자
```

```python
#4.2  독립변수들을 표준화(normalization) 한다. 
from  sklearn   import   preprocessing
X = preprocessing.StandardScaler().fit(X).transform(X)
print (X)

from  sklearn.model_selection  import   train_test_split
X_train, X_test, y_train, y_test =train_test_split( X, y, test_size=0.3, random_state = 10 ) 
# 설명:  test_size=0.3 에 의해서 7:3 비율로 훈련과 테스트를 나누고
# random_state=10 에 의해서 나중에 split 할 때도 항상 일정하게 split 할 수 있게 한다. R:seed값

print  ( 'train data 의 갯수:'  ,  X_train.shape )
print  ( 'test  data 의 갯수:'  ,  X_test.shape )
```

```python
# kNN 머신러닝 모델로 학습 시킨다
from  sklearn.neighbors  import   KNeighborsClassifier 

knn = KNeighborsClassifier( n_neighbors=5 ) 
knn.fit ( X_train, y_train )
```

```python
# 테스트 데이터에 대한 사망자와 생존자를 예측
y_hat = knn.predict( X_test ) 
```

```python
# 모델 평가를 위해 이원 교차표를 그린다
from  sklearn   import   metrics 
knn_matrix  =  metrics.confusion_matrix( y_test,  y_hat )
print(knn_matrix)
```

```python
# 모델의 정확도를 확인한다
from sklearn.metrics import accuracy_score

accuracy = accuracy_score( y_test, y_hat)
print(accuracy)
```

```python
# ■ 타이타닉 데이터를 knn 으로 수행했을때 
# 1. 필요한 패키지를 import 한다. 
import  pandas  as  pd
import  seaborn  as  sns 

#1단계.  csv ---> 데이터 프레임으로 변환 
df = sns.load_dataset('titanic')

# 컬럼이 모두다 출력될 수 있도록 출력할 열의 개수 한도를 늘리기 
pd.set_option('display.max_columns', 15 ) 

# rdf =  df.drop(['deck', 'embark_town'] , axis= 1) # 누락 데이터가 많은 column을 drop
# rdf = rdf.dropna( subset=['age'], how='any', axis= 0 ) # 나이의 누락 데이터의 행을 삭제

rdf = df

mean_age = rdf['age'].mean()
rdf['age'].fillna(mean_age, inplace=True)
rdf['deck'].fillna(rdf['deck'].value_counts(dropna=True).idxmax(), inplace=True)
rdf['embark_town'].fillna(rdf['embark_town'].value_counts(dropna=True).idxmax(),inplace=True)

most_freq = rdf['embarked'].value_counts(dropna=True).idxmax() # 정박한 항구의 누락 데이터 2건을 최빈값으로 치환
rdf['embarked'].fillna( most_freq, inplace=True ) 

ndf = rdf[ ['survived', 'pclass', 'sex', 'age', 'sibsp', 'parch', 'embarked']] 

gender = pd.get_dummies( ndf['sex'] ) # 명목형 데이터를 숫자로 변환하기 위해 더미변수를 생성
ndf = pd.concat( [ndf, gender], axis=1 ) 

onehot_embarked = pd.get_dummies( ndf['embarked'], prefix='town') # 명목형 데이터를 숫자로 변환하기 위해서 더미변수 생성
ndf = pd.concat([ ndf, onehot_embarked], axis=1)

ndf.drop( ['sex', 'embarked'],  axis=1, inplace=True ) # 명목형 column 2개를 삭제

# 훈련할 학습 데이터 X와 정답이 있는 라벨 y를 생성
X = ndf[ [ 'pclass' ,'age', 'sibsp','parch', 'female', 'male', 'town_C', 'town_Q',  'town_S']]
y = ndf['survived']   # 종속변수 

#4.2  독립변수들을 표준화(normalization) 한다. 
from  sklearn   import   preprocessing
X = preprocessing.StandardScaler().fit(X).transform(X)

from  sklearn.model_selection  import   train_test_split
X_train, X_test, y_train, y_test =train_test_split( X, y, test_size=0.3, random_state = 10 ) 
# 설명:  test_size=0.3 에 의해서 7:3 비율로 훈련과 테스트를 나누고
# random_state=10 에 의해서 나중에 split 할 때도 항상 일정하게 split 할 수 있게 한다. R:seed값

# kNN 머신러닝 모델로 학습 시킨다
from  sklearn.neighbors  import   KNeighborsClassifier 

knn = KNeighborsClassifier( n_neighbors=5 ) 
knn.fit ( X_train, y_train )

# 테스트 데이터에 대한 사망자와 생존자를 예측
y_hat = knn.predict( X_test ) 

# 모델 평가를 위해 이원 교차표를 그린다
from  sklearn   import   metrics 
knn_matrix  =  metrics.confusion_matrix( y_test,  y_hat )
print(knn_matrix)

# 모델의 정확도를 확인한다
from sklearn.metrics import accuracy_score

accuracy = accuracy_score( y_test, y_hat)
print(accuracy)
```

    나이의 결측치를 나이의 최빈값으로 치환했을 때의 정확도 : 

```python
import  pandas  as  pd
import  seaborn  as  sns 

df = sns.load_dataset('titanic')

pd.set_option('display.max_columns', 15 ) 

rdf = df

## 값 치환
rdf['age'].fillna(rdf['age'].value_counts(dropna=True).idxmax(), inplace=True) # 최빈값으로 치환
rdf['deck'].fillna(rdf['deck'].value_counts(dropna=True).idxmax(), inplace=True) # 최빈값으로 치환
rdf['embark_town'].fillna(rdf['embark_town'].value_counts(dropna=True).idxmax(),inplace=True) # 최빈값으로 치환
rdf['embarked'].fillna(rdf['embarked'].value_counts(dropna=True).idxmax() , inplace=True ) 
# rdf.isnull().sum()

## dummies
deck_dummies = pd.get_dummies(rdf['deck'], prefix='deck')
rdf = pd.concat([rdf, deck_dummies], axis=1)

gender = pd.get_dummies( rdf['sex'] ) 
rdf = pd.concat( [rdf, gender], axis=1 ) 

onehot_embarked = pd.get_dummies( rdf['embarked'], prefix='town') 
rdf = pd.concat([ rdf, onehot_embarked], axis=1)

class_dummies = pd.get_dummies(rdf['class'])
rdf = pd.concat([rdf, class_dummies], axis=1)

et_dummies = pd.get_dummies(rdf['embark_town'], prefix='embarked')
rdf = pd.concat([rdf, et_dummies], axis=1)

rdf.drop( ['sex', 'embarked', 'class', 'embark_town', 'deck'],  axis=1, inplace=True ) # 명목형 column 삭제

X = rdf[ [ 'pclass' ,'age', 'sibsp','parch', 'female', 'male', 'town_C', 'town_Q',  'town_S','embarked_Cherbourg','embarked_Queenstown','embarked_Southampton','First','Second','Third']]
y = rdf['survived']   

from  sklearn   import   preprocessing
X = preprocessing.StandardScaler().fit(X).transform(X)

from  sklearn.model_selection  import   train_test_split
X_train, X_test, y_train, y_test =train_test_split( X, y, test_size=0.3, random_state = 4 ) # random_state값 변경

from  sklearn.neighbors  import   KNeighborsClassifier 

knn = KNeighborsClassifier( n_neighbors=5 ) 
knn.fit ( X_train, y_train )

y_hat = knn.predict( X_test ) 

from  sklearn   import   metrics 
knn_matrix  =  metrics.confusion_matrix( y_test,  y_hat )
print(knn_matrix)

from sklearn.metrics import accuracy_score

accuracy = accuracy_score( y_test, y_hat)
print(accuracy)
```
