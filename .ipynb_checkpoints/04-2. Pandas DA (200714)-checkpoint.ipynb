### ※ 문제106. '시도별 전출입 인구수.xlsx' 파일을 불러와서 Pandas 데이터프레임으로 만드시오

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=(30,10)
plt.rcParams.update({'font.size':22})
census = pd.read_excel('prac/시도별 전출입 인구수.xlsx',header=0)
census
```

<b> ※ 보통 excel 데이터는 결측치 데이터가 많은데 데이터 내에 전출지별 column에 누락 데이터(NaN)이 다수 존재<br><br>
   　누락데이터는 excel 파일에서 병합된 셀을 데이터프레임으로 변환할 때 적절한 값을 찾지 못해서 발생</b>

### ※ 문제107. 누락데이터(NaN)의 앞 행을 데이터로 채우시오(excel 양식의 병합된 부분)

```python
df = census.fillna(method='ffill')
df
```

### ※ 문제108. df 데이터프레임에서 column명을 추출하시오

### ※ 문제116. 위의 데이터를 가지고 라인 그래프를 그리시오

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=(30,10)
plt.rcParams.update({'font.size':12})
pd.options.plotting.backend = "matplotlib"

census = pd.read_excel('prac/시도별 전출입 인구수.xlsx',header=0)
df = census.fillna(method='ffill')
df_seoul = df[:][(df['전출지별']=='서울특별시')&(df['전입지별']!='서울특별시')]
df_seoul.drop('전출지별',axis=1,inplace=True)
df_seoul.rename({'전입지별':'전입지'},axis=1,inplace=True)
df_seoul.set_index('전입지',inplace=True)
rs = df_seoul.loc['경기도']
plt.plot(rs.index, rs.values)
```

### ※ 문제117. $x$축 눈금의 라벨을 90도 회전 시키시오

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=(30,10)
plt.rcParams.update({'font.size':12})
pd.options.plotting.backend = "matplotlib"

census = pd.read_excel('prac/시도별 전출입 인구수.xlsx',header=0)
df = census.fillna(method='ffill')
df_seoul = df[:][(df['전출지별']=='서울특별시')&(df['전입지별']!='서울특별시')]
df_seoul.drop('전출지별',axis=1,inplace=True)
df_seoul.rename({'전입지별':'전입지'},axis=1,inplace=True)
df_seoul.set_index('전입지',inplace=True)
rs = df_seoul.loc['경기도']
plt.xticks(size=20,rotation=45)
plt.plot(rs.index, rs.values)
```
