### <b> ■ 결측치 확인을 함수로 만들어서 확인하는 방법</b>
#### 예제 : kaggle 타이타닉 데이터프레임 전체의 결측치를 확인하는 함수 생성

```python
import pandas as pd
df = pd.read_csv('train.csv')

def missing_values(x):
    return x.isnull()

# print(missing_values(df))

def missing_counts(x):
    return missing_values(x).sum()

print(missing_counts(df))
```

### ※ 문제184. 아래의 결측치 값을 다 더하는 함수를 total_number_missing으로 생성하시오
    앞에서 만든 missing_counts를 활용하시오

```python
def total_number_missing(x):
    return sum(missing_counts(x))

total_number_missing(df)
```

#### <b> ※ 데이터프레임에 함수를 매핑하는 문법 2가지</b>
    1. 데이터프레임객체.apply(매핑함수명)
    2. 데이터프레임객체.pipe(매핑함수명)

```python
df.pipe(total_number_missing), df.pipe(missing_counts), df.pipe(missing_values)
```

### <b>■ 열(column) 분리</b>
    하나의 열이 여러가지 정보를 담고 있을 때 각 정보를 서로 분리해서 사용하는 경우
    예시
        '년월일' 정보가 있을 때 '년', '월', '일' 로 구분해서 3개의 열로 만든다.
        사람의 이름이 들어있는 열을 '성'과 '이름'으로 구분해서 만든다.
#### <b> 주가데이터.xlsx에서 '년월일' column을 3개로 분리하는 예제</b>

```python
import pandas as pd
df = pd.read_excel('prac/주가데이터.xlsx')
df.head()
```

```python
df['연월일'] = df['연월일'].astype('str') # 문자열로 변환
dates = df['연월일'].str.split('-')
dates
```

```python
df['연'] = dates.str.get(0)
df['월'] = dates.str.get(1)
df['일'] = dates.str.get(2)
df.head()
```

### ※ 문제195. kaggle 타이타닉 나이의 결측치를 해당 호칭의 평균값으로 치환하시오

```python
import pandas as pd
tit = pd.read_csv('train.csv')
appel = tit['Name'].str.split(',').str[1].str.strip().str.split('.')
tit['title'] = appel.str.get(0)
age_m = pd.DataFrame(round(tit.groupby('title')['Age'].mean(),1))
rs = pd.merge(tit, age_m, on='title',how='left')
tit = rs
tit.Age_x.fillna(tit.Age_y, inplace=True)
tit.head(30)
```

    Kaggle 4% 안에 들어가려면 파생변수 2개가 필요
        1. 호칭을 이용해서 나이의 결측치를 채워넣는다
        2. 여자와 아이는 1로 나머지는 0으로 하는 파생변수
