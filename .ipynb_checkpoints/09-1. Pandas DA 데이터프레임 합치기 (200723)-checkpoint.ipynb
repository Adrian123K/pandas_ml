### <b>■ pandas.concat</b>
    문법
        pd.concat([df1,df2],옵션)
        옵션
            axis=0, 1 (행, 열 방향)
            ignor_index = True(행 번호 새로 생성)
            join='inner' (column들의 교집합이 출력)
            join='outer' (모든 column들이 다 출력)

````python
# 예제1. seaborn에서 제공하는 타이타닉 데이터와 Kaggle에서 제공하는 타이타닉 데이터가 서로 같은 데이터인지 확인
import seaborn as sns
import pandas as pd

df1 = sns.load_dataset('titanic')
df2 = pd.read_csv('train.csv')
len(df1), len(df2)
````

````python
# 예제2. seaborn의 타이타닉 나이 데이터와 Kaggle의 타이타닉 나이 데이터를 서로 연결
df1['age'], df2['Age']
pd.concat([df1.age,df2.Age],axis=1)
````

````python
# 예제3. df1, df2 concat으로 연결(카페 스크립트 자료)
import pandas as pd

# 데이터프레임 만들기
df1 = pd.DataFrame({'a': ['a0', 'a1', 'a2', 'a3'],
                    'b': ['b0', 'b1', 'b2', 'b3'],
                    'c': ['c0', 'c1', 'c2', 'c3']},
                   index=[0, 1, 2, 3])

df2 = pd.DataFrame({'a': ['a2', 'a3', 'a4', 'a5'],
                    'b': ['b2', 'b3', 'b4', 'b5'],
                    'c': ['c2', 'c3', 'c4', 'c5'],
                    'd': ['d2', 'd3', 'd4', 'd5']},
                   index=[2, 3, 4, 5])

print(df1, '\n')
print(df2, '\n')
````

````python
rs = pd.concat([df1,df2],axis=1,join='outer')
rs2 = pd.concat([df1,df2],axis=1,join='inner')
rs, rs2
````

### ※ 문제201. emp와 dept를 merge를 사용해서 조인하시오
    조인해서 이름과 부서위치를 출력하시오

````python
import pandas as pd
emp = pd.read_csv('emp.csv')
dept = pd.read_csv('dept.csv')

df = pd.merge(emp, dept, on='deptno', how='inner')
df[['ename','loc']]
````

### <b>■ Pandas merge 문법</b>
    문법
        pandas.merge(df1, df2, on='', how='')
        옵션
            how = 'inner' (열의 데이터가 양쪽 데이터프레임에 공통으로 존재하는 교집합만 추출)
            how = 'outer' (열의 데이터가 양쪽 데이터프레임에 공통으로 존재하는 교집합이 아니어도 추출)
            how = 'left' (왼쪽 데이터프레임의 키 열에 속하는 데이터 값을 기준으로 병합)
            how = 'right' (오른쪽 데이터프레임의 키 열에 속하는 데이터 값을 기준으로 병합)
            

### ※ 문제202. 아래의 SQL을 Pandas로 구현하시오
```sql
select e.ename, d.loc
    from emp e, dept d
    where e.deptno(+) = d.deptno;
```

````python
import pandas as pd
emp = pd.read_csv('emp.csv')
dept = pd.read_csv('dept.csv')

df = pd.merge(emp, dept, how='right',on='deptno')
df[['ename','loc']]
````

### ※ 문제207. 아래의 SQL을 위의 조인방법으로 구현하시오
```sql
select e.ename, d.loc
    from emp e full outer join dept d
    on (e.deptno = d.deptno);
```

````python
import pandas as pd
emp = pd.read_csv('emp.csv', index_col='deptno')
dept = pd.read_csv('dept.csv', index_col='deptno')
df = emp.join(dept,how='outer')
df[['ename','loc']]
````

### <b>■ 그룹연산</b>
    복잡한 데이터를 어떤 기준에 따라 여러 그룹으로 나눠서 관찰하는 것도 좋은 방법인데 이처럼 특정 기준을 적용하여 몇 개의 그룹으로 분할하여 처리하는 것
    그룹연산의 3가지 단계
        1단계 : 분할(split)
            데이터를 특정 조건에 의해 분할
        2단계 : 적용(apply)
            데이터 집계, 변환, 필터링하는데 필요한 메소드
        3단계 : 결합(combine)
            2단계의 처리를 하나로 결합

#### <b>■ 1단계. 분할</b>
    문법
        df.groupby(기준 열)


````python
# 예제1. 운임 클래스 first, second, third 중에 생존율이 높은 클래스는?
import seaborn as sns
tit = sns.load_dataset('titanic')
df = tit.loc[:,['age','sex','class','fare','survived']]

grouped = df.groupby('class')

for key, group in grouped:
    print(f"key: {key}\n number:{len(group)}\n {group.head()}\n")
````

````python
# 생존율 평균 확인
average = grouped.mean()
average
````

### ※ 문제208. 사원 테이블을 위와 같이 분할하여 부서번호별로 월급의 평균값들이 출력되게하고 부서번호별로 월급의 평균이 높은 부서번호가 어딘지 확인하시오

````python
import pandas as pd
emp = pd.read_csv('emp.csv')
df = emp.loc[:,['deptno','ename','sal']]
gp = df.groupby(['deptno'])

for key, v in gp:
    print(f"key:{key}\n number:{len(v)}\n {v.head()}\n")
````

````python
gp_avg = gp.mean()
gp_avg
````

### ※ 문제209. 운임별 클래스별로는 first클래스가 가장 높은 생존율을 보인 것을 확인했다.그러면 클래스별로 남자와 여자 중에 더 생존율이 높은 성별은 무엇인가

````python
import seaborn as sns
import pandas as pd
tit = sns.load_dataset('titanic')
df = tit.loc[:,['age','sex','class','fare','survived']]

grouped = df.groupby(['class','sex'])

for key, group in grouped:
    print(f"key: {key}\n number:{len(group)}\n {group.head()}\n")
    
avg = grouped.mean()
avg
````

### <b>■ 2단계와 3단계의 적용 및 결합 단계</b>
    앞에서 분할한 그룹 객체에 대하여 각 그룹별 평균을 계산한 것 처럼 그룹 객체에 다양한 연산을 적용할 수 있다.

    집계 기능을 내장하고 있는 Pandas 기본 함수에는 mean, max, min, sum, count, var, std, describe, info, size, first, last 등이 있다.

### ※ 문제211. 요금을 나타내는 fare 열은 first, second, third 클래스 중에 요금의 표준편차가 큰 클래스는 어디인가

````python
import seaborn as sns
import pandas as pd
tit = sns.load_dataset('titanic')
df = tit.loc[:,['class','fare']]
gp = df.groupby(['class'])

std = gp.std()
std
````
